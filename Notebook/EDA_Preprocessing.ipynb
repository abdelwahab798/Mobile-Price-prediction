```python
print(df.duplicated().sum())
print("*"*30)
df.isna().sum()
```

```python
df["gpu_score"].describe()
```

```python
df=df.loc[df["gpu_score"]>0]
```

```python
df["gpu_score"].describe()
```

```python
df["brand"].unique()
```

```python
df.loc[df["brand"]=="Nothing"]=np.nan
df.dropna(inplace=True)
```

```python
sns.countplot(df["brand"])
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=45)
sns.boxenplot(data=df[["price_inr", "gpu_score", "cpu_score", "battery_mah"]])
```

```python
most_expensive=df.loc[df["price_inr"]>125000]
print(f"number of expensive {most_expensive.shape[0]}")
sns.countplot(x=most_expensive["brand"])
plt.title("Brand of most expensive mobiles")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price_inr'],kde=True,color='blue')
plt.title('Distribution of Smartphone Prices')
plt.xlabel('Price (INR)')
```

```python
brand_price=df.groupby('brand')['price_inr'].mean().sort_values(ascending=False)
sns.barplot(x=brand_price.index, y=brand_price.values, order=brand_price.index)
plt.xticks(rotation=45)
plt.title('Price Range per Brand')
```

```python
correlation = df.select_dtypes(include=['number']).corr()['price_inr'].sort_values(ascending=False)
plt.figure(figsize=(10, 8))
sns.barplot(x=correlation.values, y=correlation.index, palette='viridis')
plt.title('Correlation of Features with Price')
plt.xlabel('Correlation Coefficient')
```

```python
sns.scatterplot(data=df, x='gpu_score', y='price_inr')
plt.title('Relationship between GPU Score and Price')
```

```python
sns.scatterplot(data=df, x='cpu_score', y='price_inr')
plt.title('Relationship between GPU Score and Price')
```

```python
plt.figure(figsize=(15, 8))
sns.violinplot(data=df,x='brand',y='price_inr')
plt.xticks(rotation=45)
plt.title('Price Distribution Density per Brand')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df,x='display_type',y='price_inr',hue='refresh_rate_hz')
plt.title('Average Price by Display Type and Refresh Rate')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='brand', y='battery_mah', estimator='mean')
plt.title('Average Battery Capacity')
```

```python
df = df.drop(columns=[
    "display_size_inch",
    "bluetooth_version",
    "dual_sim",
    "screen_to_body_ratio",
    "water_resistance",
    "5g_support",
    "face_unlock",
    "front_camera_mp",
    "warranty_years",
    "colors_available",
    "storage_gb",
    "weight_g",
    "launch_year",
    "fast_charging_w",
    "wireless_charging",
    "fingerprint_sensor",
    "model_name",
    "os"
], errors="ignore")
```
